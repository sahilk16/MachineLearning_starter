```python
import numpy as np
import pandas as pd
% matplotlib inline
import matplotlib.pyplot as plt
```

```python
first = pd.read_csv("E:/Projects/DeepLearning/churn.csv")
first.head()
```

```python
first.info()
```

```python
#dropping the unwanted variables from the dataset
first.drop(["State"], axis = 1, inplace = True)
first.drop(["Phone"], axis = 1, inplace = True)
first.drop(["Area Code"], axis = 1, inplace = True)
```

```python
#As the dataset contains the missing value in the form of "?", we need to replace it by 0
first['Day Charge'].replace("?", np.NaN, inplace = True)
first['Eve Mins'].replace("?", np.NaN, inplace = True)
first['Eve Calls'].replace("?", np.NaN, inplace = True)
first['Night Charge'].replace("?", np.NaN, inplace = True)
first['Intl Calls'].replace("?", np.NaN, inplace = True)
first['Intl Charge'].replace("?", np.NaN, inplace = True)
first["Int'l Plan"].replace("?", np.random.choice(['yes', 'no']), inplace = True)
first["VMail Plan"].replace("?", np.random.choice(['yes', 'no']), inplace = True)
```

```python
#Changing the datatypes of all the variables accordingly - float or int
first["Day Charge"] = first['Day Charge'].astype(str).astype(float)
first["Eve Mins"] = first['Eve Mins'].astype(str).astype(float)
first["Eve Calls"] = first['Eve Calls'].astype(str).astype(float)
first["Night Charge"] = first['Night Charge'].astype(str).astype(float)
first["Intl Calls"] = first['Intl Calls'].astype(str).astype(float)
first["Intl Charge"] = first['Intl Charge'].astype(str).astype(float)
```

```python
first.info()
```

```python
#replacing the NaN values with mean of their values
first['Day Charge'].replace(np.NaN, first["Day Charge"].mean(), inplace = True)
first['Eve Mins'].replace(np.NaN,first["Eve Charge"].mean(), inplace = True)
first['Night Charge'].replace(np.NaN,first["Night Charge"].mean(), inplace = True)
first['Intl Charge'].replace(np.NaN, first["Intl Charge"].mean(), inplace = True)
```

```python
first.info()
```

```python
var = ["Int'l Plan", "VMail Plan","Eve Calls", "Intl Calls", "Churn?" ]
for v in var:
    print("Frequency count in these variables: ")
    print(first[v].value_counts())
```

```python
first["Eve Calls"].replace(np.NaN, first["Eve Calls"].median(), inplace = True)
first["Intl Calls"].replace(np.NaN, first["Intl Calls"].median(), inplace = True)
```

```python
first.info()
```

```python
# Isolate target data
churn_result = first['Churn?']
y = np.where(churn_result == 'True.',1,0)
```

```python
first.drop(["Churn?"], axis = 1, inplace = True)
```

```python
first.head()
```

```python
yes_no_cols = ["Int'l Plan","VMail Plan"]
first[yes_no_cols] = first[yes_no_cols] == 'yes'
```

```python
# Lets put in some new features in the dataset - Feature Engineering

first["Total Charges"] = first["Day Charge"] + first["Eve Charge"] + first["Night Charge"] + first["Intl Charge"]
```

```python
from sklearn.preprocessing import StandardScaler

X = first.as_matrix().astype(np.float)

# This is important
scaler = StandardScaler()
X = scaler.fit_transform(X)
print ("Feature space holds %d observations and %d features" % X.shape)
print ("Unique target labels:", np.unique(y))
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier

kf = KFold(n_splits = 10, shuffle = True, random_state = 42)
model = RandomForestClassifier()
ypred=np.zeros_like(y,dtype=float)

def run(X, y, model, n_estimators):
    for train_index, test_index in kf.split(X):
        #print("TRAIN:", train_index, "TEST:", test_index)
        x_train, y_train = X[train_index], y[train_index]
        x_test, y_test = X[test_index], y[test_index]
        
        model.fit(x_train, y_train)
        
        ypred[test_index]= model.predict_proba(x_test)[:,1]
        
        
    return ypred

run(X, y , model, n_estimators = 10)

score = roc_auc_score(y, ypred)
score
```
